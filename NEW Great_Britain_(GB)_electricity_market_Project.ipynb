 **3.Trading Strategy Proposal:**


*  Proposed Strategy:

    Buy at Min Price and Sell at Max Price:
    For each trading day, buy the asset at the Min Price and sell it at the Max Price

    Take-Profit at Max Price:
    Sell when the price reaches the Max Price for the day, as it represents the highest price observed for the day.

    

*  Rationale for Strategy:

    Min-Max Strategy: This is a straightforward approach of taking advantage of daily price fluctuations. If the min and max prices follow a consistent pattern during the day, buying low and selling high can yield predictable returns.

    Regular Patterns in Min and Max Times: The table shows that min prices tend to occur early in the day , while max prices generally occur later in the afternoon or evening. This suggests a potential trend of rising prices throughout the day, which the strategy capitalizes on.





*  **Assumptions:**

    Market Efficiency: Prices move predictably within the identified min and max times.

    No Overnight Holding: The strategy assumes that no positions are carried overnight, minimizing the risk of large gaps in prices between trading days.





*   **Profitability Estimate:**

    Using the Min Price and Max Price data for each day, we can calculate the profit for each day by subtracting the Min Price from the Max Price.

   
    Formuala for daily profit=(Max price-Min price)
